```python
# PEGANDO DADOS DA PLANILHA
import pandas as pd

sheet_name = 'data_example.csv'
print(f'-> Importando dados da planilha "{sheet_name}" ...')
df = pd.read_csv(sheet_name).dropna(axis='index', how='all')
print("ARQUIVO IMPORTADO COM SUCESSO!")
print(f"- Qtd perguntas: {df.shape[1]-1}")
print(f"- Qtd avaliadores: {df.shape[0]-1}")
```

```python
# IMPORTING AND FORMATTING STUFF
from IPython.display import SVG, HTML
from pygal.style import Style
import pygal

html_pygal = """
<!DOCTYPE html>
<html>
  <head>
  <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js"></script>
  <script type="text/javascript" src="https://kozea.github.io/pygal.js/2.0.x/pygal-tooltips.min.js"></script>
  <script type="text/javascript" src="https://rawgit.com/s3rvac/pygal-toggle-graphs/master/pygal-toggle-graphs.js"></script>
  </head>
  <body>
      {pygal_render}
  </body>
</html>
"""

color_pallete = (
    '#0080FF', # Azure Blue
    '#000080', # Navy Blue
    '#111E6C', # Royal Blue
)

custom_style = Style(
  background='transparent',
  plot_background='#ffffff',
  foreground='#636363',
  foreground_subtle='#adadad',
  foreground_strong='#000000',
  guide_stroke_dasharray = '0',
  major_guide_stroke_dasharray = '0',
  opacity='.3',
  opacity_hover='.9',
  transition='100ms ease-in',
  title_font_size = 20,
  label_font_size = 14,
  tooltip_font_size = 14,
  colors=color_pallete,
)

def generate_visualization(data, title='Sem nome'):
    radar_chart = pygal.Radar(
        width=870,
        height=750,
        fill=True,
        style=custom_style,
        truncate_label=2,
        dots_size=3,
        tooltip_border_radius=5,
        stroke_style={'width': 3, 'linecap': 'round', 'linejoin': 'round'}
    )
    radar_chart.title = title
    radar_chart.x_labels = data.columns[1:]
    radar_chart.y_labels = [0, 1, 2, 3, 4 , 5]
    for values in data.values:
        avaliador = values[0]
        avaliacao = values[1:]
        radar_chart.add(avaliador, avaliacao)
    display(HTML(html_pygal.format(pygal_render=radar_chart.render(is_unicode=True))))

    
# ---------------------
# GERANDO VISUALIZAÇÃO   
# ---------------------

graph_name = "Avaliação das minhas competências"
generate_visualization(data=df, title=graph_name)
```
